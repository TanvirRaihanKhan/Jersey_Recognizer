```python
#!export
model=load_learner((f"checkpoints/models/jersey-recogniser-v2.pkl"))
```

```python
#!export
jersey_labels=[
    'AC Milan Jersey',
    'Arsenal Jersey',
    'Atletico Madrid Jersey',
    'Barcelona Jersey',
    'Bayern Munich Jersey',
    'Borussia Dortmund Jersey',
    'Chelsea Jersey',
    'FC Porto Jersey',
    'Inter Milan Jersey',
    'Juventus Jersey',
    'Leeds United Jersey',
    'Liverpool Jersey',
    'Manchester City Jersey',
    'Manchester United Jersey',
    'Napoly Jersey',
    'Newcastle Jersey',
    'PSG Jersey',
    'Real Madrid Jersey',
    'Southampton Jersey',
    'Tottenham Jersey'
]

def recognize_image(image):
  pred,idx,probs=model.predict(image)
  print(pred)
  return dict(zip(jersey_labels, map(float,probs)))

```

```python

img=PILImage.create(f"Test Data/Test_image1.jpg")
img.thumbnail((128,128))
img
```

```python
recognize_image(img)
```

```python
#!export
import gradio as gr

image = gr.inputs.Image(shape=(192,192))
label = gr.outputs.Label()
examples = [
   "Test Data/Test_image1.jpg",
   "Test Data/Test_image2.jpg",
   "Test Data/Test_image3.jpg",
   "Test Data/Test_image4.jpg"]

iface = gr.Interface(fn=recognize_image, inputs=image, outputs=label,examples=examples)
iface.launch(inline=False,share=True)
```
